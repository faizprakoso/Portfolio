   - discounted price
   - actual price
   - discount percentage
   - rating

### Handle Null Value

```python
df[df['rating_count'].isnull()]
```

```python
clean_data=(df['rating_count'][df['rating_count'].notnull()].astype(str).replace(',','',regex=True)).astype(int)
sns.histplot(clean_data,bins=30)
```

```python
median=(df['rating_count'][df['rating_count'].notnull()].astype(str).replace(',','',regex=True)).astype(int).median()
```

```python
df['rating_count']=df['rating_count'].fillna(median)
```

### Change Data Type

```python
# make sure every price does not contain '‚Çπ,' and make it into integer
df['discounted_price'] = df['discounted_price'].astype(str).replace('[‚Çπ,]', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].astype(str).replace('[‚Çπ,]', '', regex=True).astype(float)
df['rating_count'] = df['rating_count'].astype(str).replace('[,]', '', regex=True).astype(float)

# remove the '%' from the discount and change the data into float and divide it by 100
df['discount_percentage'] = df['discount_percentage'].astype(str).replace('[%]', '', regex=True).astype(float)/100
```

```python
# The rating data type can't be change and apparently there is other character involve, which is |
df['rating'].value_counts()
```

```python
# we need to change it into number. One of kaggle user has found the actual number of the product is 3.9
df['rating']=df['rating'].astype(str).replace('\|', '3.9', regex=True).astype(float)
```
